```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

%matplotlib inline
sns.set_style('darkgrid')
plt.rcParams['font.size'] = 14
plt.rcParams['figure.figsize'] = (17, 5)
plt.rcParams['figure.facecolor'] = '#00000000'
```

```python
# Load the COVID-19 dataset
df = pd.read_csv('https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv')

# Select relevant columns
rnr = df[['continent', 'location', 'total_cases', 'total_cases_per_million', 'total_deaths', 'total_deaths_per_million', 'people_vaccinated', 'people_fully_vaccinated', 'total_boosters', 'people_vaccinated_per_hundred', 'people_fully_vaccinated_per_hundred', 'total_boosters_per_hundred', 'population', 'population_density', 'median_age', 'aged_65_older', 'aged_70_older', 'gdp_per_capita', 'life_expectancy', 'human_development_index']]

# Group by location and select maximum values for each metric
dfe = rnr.groupby('location')[['continent', 'total_cases', 'total_cases_per_million', 'total_deaths', 'total_deaths_per_million', 'people_vaccinated', 'people_fully_vaccinated', 'total_boosters', 'people_vaccinated_per_hundred', 'people_fully_vaccinated_per_hundred', 'total_boosters_per_hundred', 'population', 'population_density', 'median_age', 'aged_65_older', 'aged_70_older', 'gdp_per_capita', 'life_expectancy', 'human_development_index']].max()

# Reset index for easier manipulation
sd = dfe.reset_index()

# Remove rows with unwanted indices
sm = sd.drop(index=[1, 12, 70, 71, 96, 127, 128, 162, 169, 211, 241, 251])

# Reset index and sort values for further analysis
sn = sm.reset_index().sort_values(['continent', 'location']).reset_index()

# Extract and plot Top 20 Countries by Total Cases
top20_ca = sn.sort_values('total_cases', ascending=False).head(20)
top20_cases = top20_ca[['location', 'total_cases']].reset_index().drop('index', axis=1)
```

```python
sns.barplot(x='total_cases', y='location', data=top20_cases, palette='viridis')
plt.title('Top 20 Countries by Total Cases')
plt.xlabel('Total Cases')
plt.ylabel('Country')
plt.xticks(rotation=90)
plt.show()
```

```python
# Extract and plot Top 20 Countries by Total Cases Per Million
top20_capm = sn.sort_values('total_cases_per_million', ascending=False).head(20)
top20_casespm = top20_capm[['location', 'total_cases_per_million']].reset_index().drop('index', axis=1)

sns.barplot(x='total_cases_per_million', y='location', data=top20_casespm, palette='magma')
plt.title('Top 20 Countries by Total Cases Per Million')
plt.xlabel('Total Cases Per Million')
plt.ylabel('Country')
plt.xticks(rotation=90)
plt.show()
```

```python
# Extract and plot Top 20 Countries by Total Deaths
top20_de = sn.sort_values('total_deaths', ascending=False).head(20)
top20_deaths = top20_de[['location', 'total_deaths']].reset_index().drop('index', axis=1)

sns.barplot(x='total_deaths', y='location', data=top20_deaths, palette='plasma')
plt.title('Top 20 Countries by Total Deaths')
plt.xlabel('Total Deaths')
plt.ylabel('Country')
plt.xticks(rotation=90)
plt.show()
```

```python
# Extract and plot Top 20 Countries by Total Deaths Per Million
top20_depm = sn.sort_values('total_deaths_per_million', ascending=False).head(20)
top20_deathspm = top20_depm[['location', 'total_deaths_per_million']].reset_index().drop('index', axis=1)

sns.barplot(x='total_deaths_per_million', y='location', data=top20_deathspm, palette='inferno')
plt.title('Top 20 Countries by Total Deaths Per Million')
plt.xlabel('Total Deaths Per Million')
plt.ylabel('Country')
plt.xticks(rotation=90)
plt.show()
```

```python
# Extract and plot Top 20 Countries by People Vaccinated
top20_vac = sn.sort_values('people_vaccinated', ascending=False).head(20)
top20_vaccinated = top20_vac[['location', 'people_vaccinated']].reset_index().drop('index', axis=1)

sns.barplot(x='people_vaccinated', y='location', data=top20_vaccinated, palette='coolwarm')
plt.title('Top 20 Countries by People Vaccinated')
plt.xlabel('People Vaccinated')
plt.ylabel('Country')
plt.xticks(rotation=90)
plt.show()
```

```python
# Extract and plot Top 20 Countries by People Fully Vaccinated
top20_fullvac = sn.sort_values('people_fully_vaccinated', ascending=False).head(20)
top20_fullvaccinated = top20_fullvac[['location', 'people_fully_vaccinated']].reset_index().drop('index', axis=1)

sns.barplot(x='people_fully_vaccinated', y='location', data=top20_fullvaccinated, palette='seismic')
plt.title('Top 20 Countries by People Fully Vaccinated')
plt.xlabel('People Fully Vaccinated')
plt.ylabel('Country')
plt.xticks(rotation=90)
plt.show()
```

```python
# Extract and plot Top 20 Countries by Total Boosters
top20_boos = sn.sort_values('total_boosters', ascending=False).head(20)
top20_boosters = top20_boos[['location', 'total_boosters']].reset_index().drop('index', axis=1)

sns.barplot(x='total_boosters', y='location', data=top20_boosters, palette='copper')
plt.title('Top 20 Countries by Total Boosters')
plt.xlabel('Total Boosters')
plt.ylabel('Country')
plt.xticks(rotation=90)
plt.show()
```

```python
# Plot a bar chart for the Top 20 Countries by Total Boosters Per Hundred
top20_boos_per_hundred = sn.sort_values('total_boosters_per_hundred', ascending=False).head(20)
sns.barplot(x='total_boosters_per_hundred', y='location', data=top20_boos_per_hundred, palette='plasma')
plt.xlabel('Total Boosters Per Hundred')
plt.ylabel('Location')
plt.title('Top 20 Countries by Total Boosters Per Hundred')
plt.show()
```

```python
# Fill missing values in 'continent' with 'Unknown'
sn['continent'] = sn['continent'].fillna('Unknown')

# Plot Sunburst chart for Total Cases Over Time
fig = px.sunburst(sn, path=['continent', 'location'], values='total_cases')
fig.update_layout(width=900, height=800, title='Total Cases Over Time', title_x=0.5)
fig.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 6))

# Plot bar chart for Total Cases by Continent
sns.barplot(x='total_cases', y='continent', data=df, palette='plasma', ax=axs[0])
axs[0].set_xlabel('Total Cases')
axs[0].set_ylabel('Continent')
axs[0].set_title('Total Cases by Continent')

# Plot bar chart for Total Deaths by Continent
sns.barplot(x='total_deaths', y='continent', data=df, palette='plasma', ax=axs[1])
axs[1].set_xlabel('Total Deaths')
axs[1].set_ylabel('Continent')
axs[1].set_title('Total Deaths by Continent')

# Adjust layout
plt.tight_layout()

# Show the subplots
plt.show()
```

```python
# Create a 1x2 grid of subplots
fig, axs = plt.subplots(1, 2, figsize=(15, 6))

# Plot bar chart for People Vaccinated by Continent
sns.barplot(x='continent', y='people_vaccinated', data=df, palette='plasma', ax=axs[0])
axs[0].set_xlabel('Continent')
axs[0].set_ylabel('People Vaccinated')
axs[0].set_title('People Vaccinated by Continent')
axs[0].tick_params(axis='x', rotation=45)

# Plot bar chart for People Fully Vaccinated by Continent
sns.barplot(x='continent', y='people_fully_vaccinated', data=df, palette='plasma', ax=axs[1])
axs[1].set_xlabel('Continent')
axs[1].set_ylabel('People Fully Vaccinated')
axs[1].set_title('People Fully Vaccinated by Continent')
axs[1].tick_params(axis='x', rotation=45)

# Adjust layout
plt.tight_layout()

# Show the subplots
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 6))

# Plot bar chart for Population by Continent
sns.barplot(x='continent', y='population', data=df, palette='plasma', ax=axs[0])
axs[0].set_xlabel('Continent')
axs[0].set_ylabel('Population')
axs[0].set_title('Population by Continent')
axs[0].tick_params(axis='x', rotation=45)

# Plot bar chart for Population Density by Continent
sns.barplot(x='continent', y='population_density', data=df, palette='plasma', ax=axs[1])
axs[1].set_xlabel('Continent')
axs[1].set_ylabel('Population Density')
axs[1].set_title('Population Density by Continent')
axs[1].tick_params(axis='x', rotation=45)

# Adjust layout
plt.tight_layout()

# Show the subplots
plt.show()
```
